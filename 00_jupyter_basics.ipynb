# Jupyter basics - Magic Functions

_Reference: [ipython](https://ipython.readthedocs.io/en/stable/interactive/magics.html)_

To Jupyter users: Magics are specific to and provided by the IPython kernel. Whether Magics are available on a kernel is a decision that is made by the kernel developer on a per-kernel basis. To work properly, Magics must use a syntax element which is not valid in the underlying language. For example, the IPython kernel uses the <code>%</code> syntax element for Magics as <code>%</code> is not a valid unary operator in Python. However, <code>%</code> might have meaning in other languages.

# %alias

Define an alias for a system command.

‘%alias alias_name cmd’ defines ‘alias_name’ as an alias for ‘cmd’

Then, typing ‘alias_name params’ will execute the system command ‘cmd params’ (from your underlying operating system).

Aliases have lower precedence than magic functions and Python normal variables, so if ‘foo’ is both a Python variable and an alias, the alias can not be executed until ‘del foo’ removes the Python variable.

You can use the %l specifier in an alias definition to represent the whole line when the alias is called.

```python
%alias bracket echo "Input in brackets: <%l>"
```
